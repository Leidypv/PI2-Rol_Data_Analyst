####Dataframe con datos históricos del mercado que incluyen precio, capitalización de mercado y volumen de 24 horas.       
#### ***Criptomonedas***: bitcoin, binancecoin, cardano, chives-coin, dogecoin, ethereum, gala, ripple, solana, terra-luna.
#### ***Monedas de cambio***: usd.

```python
import datetime
import pandas as pd
from pycoingecko import CoinGeckoAPI

# Crear una instancia de la API de CoinGecko
cg = CoinGeckoAPI()

# Rango de tiempo en UNIX Timestamp
desde = 1577836800  # 01/01/2020
hasta = 1672531200  # 01/01/2023

# Criptoactivos a consulta
id = ['bitcoin', 'binancecoin', 'cardano', 'chives-coin', 'dogecoin', 'ethereum', 'gala', 'ripple', 'solana','terra-luna']

# Monedas de cambio en las que deseo obtener los precios
change_currency = ['usd']

# Recorrer las criptomonedas y las monedas de cambio
resultados = []
for cripto in id:
    for moneda in change_currency:
        # Realizar la consulta histórica
        historical_data = cg.get_coin_market_chart_range_by_id(id=cripto, vs_currency=moneda, from_timestamp=desde, to_timestamp=hasta)

        # Extraer los precios de la respuesta
        prices = historical_data['prices']
        # Crear dataFrame con los precios
        df = pd.DataFrame(prices, columns=['date', 'price'])
        df['date'] = pd.to_datetime(df['date'], unit='ms')

        # Agregar la información de criptomoneda y moneda de cambio como columnas
        df['id'] = cripto
        df['change_currency'] = moneda

        # Agregar
        resultados.append(df)

# Concatenar los DataFrames individuales en uno solo
dataframe = pd.concat(resultados, ignore_index=True)

# conversión Dtype
dataframe['date']= pd.to_datetime(dataframe['date'])
dataframe['id']= dataframe['id'].astype('category')
dataframe['change_currency']= dataframe['change_currency'].astype('category')
dataframe
```

###Dataframe datos del mercado criptomoneda en usd.

```python
import pandas as pd
import numpy as np
import requests
from pandas import json_normalize
import warnings
warnings.filterwarnings('ignore')

url_query='https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=100&page=1&sparkline=false&locale=en'
response=requests.get(url_query, params={'format':'json'})
response.content

# Desanidar diccionarios
data=response.json()
data=pd.json_normalize(data)
market1 = data[['id','current_price','market_cap_rank','market_cap','total_volume','circulating_supply','total_supply','max_supply']]

# Selecciono las filas que necesito
market1 = market1.loc[market1['id'].isin(['bitcoin', 'binancecoin', 'cardano', 'chives-coin', 'dogecoin', 'ethereum', 'gala','ripple','solana', 'terra-luna', ])]
market1
```

```python
url_query='https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&ids=chives-coin&order=market_cap_desc&per_page=100&page=1&sparkline=false&locale=en'
response=requests.get(url_query, params={'format':'json'})
response.content

# Desanidar diccionarios
data=response.json()
data=pd.json_normalize(data)

mark_et = data[['id','current_price','market_cap_rank','market_cap','total_volume','circulating_supply','total_supply','max_supply']]
mark_et
```

```python
market = [market1,mark_et]
market = pd.concat(market,ignore_index=True)
market
```

```python
# Reemplazar los valores none por 0 de la fila
fila_index = 1
market.iloc[fila_index] = market.iloc[fila_index].fillna(0)
fila_index = 4
market.iloc[fila_index] = market.iloc[fila_index].fillna(0)
fila_index = 6
market.iloc[fila_index] = market.iloc[fila_index].fillna(0)
fila_index = 8
market.iloc[fila_index] = market.iloc[fila_index].fillna(0)
fila_index = 9
market.iloc[fila_index] = market.iloc[fila_index].fillna(0)

# Formato decimal
columns_to_format = ['current_price','market_cap','total_volume','circulating_supply','total_supply','max_supply']
market = market[['id','market_cap_rank'] + columns_to_format]
market[columns_to_format] = market[columns_to_format].applymap(lambda x: format(x, '.3f'))
market
```

```python
# Cambiar type
market['id']= market['id'].astype('category')
market['market_cap_rank']= market['market_cap_rank'].astype('int')
convert =['current_price','market_cap','total_volume','circulating_supply','total_supply','max_supply']
market[convert]= market[convert].astype('float')
market
```
